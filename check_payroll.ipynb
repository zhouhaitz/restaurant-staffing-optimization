```python
import pandas as pd
```

```python
payroll = pd.read_csv("PayrollExport_2025_10_01-2025_11_30 (1).csv")
```

```python
avg_hourly_by_job = payroll.groupby('Job Title')['Hourly Rate'].mean()

roles_of_interest = [
    "Server",
    "Line Cook",
    "Host",
    "Runner",
    "Busser"
]

filtered_hourly_by_job = avg_hourly_by_job[avg_hourly_by_job.index.isin(roles_of_interest)]
filtered_hourly_by_job
```
