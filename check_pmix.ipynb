```python
import pandas as pd

pmix = pd.read_csv('Product Mix 10.1 to 11.30.25.csv')
pmix
```

```python
dinner_pmix = pmix[pmix['Menu'] == 'DINNER']
# Fix column name to match the CSV ("Qty sold" instead of "Qty Sold")
clean_dinner_pmix = dinner_pmix[['Menu', 'Menu group', 'Item, open item', 'Sales Category', 'Qty sold']].copy()

#drop duplicates
clean_dinner_pmix = clean_dinner_pmix.drop_duplicates()

#rename columns
clean_dinner_pmix.rename(columns={'Item, open item': 'Item'}, inplace=True)

#drop last 3 rows
clean_dinner_pmix = clean_dinner_pmix.iloc[:-3]

days_open = 60
#daily average per product
clean_dinner_pmix['Daily Avg Qty Sold'] = clean_dinner_pmix['Qty sold'] / days_open

clean_dinner_pmix
```

```python
clean_dinner_pmix.dropna(inplace=True)
clean_dinner_pmix.sort_values(by='Daily Avg Qty Sold', ascending=False, inplace=True)

clean_dinner_pmix[clean_dinner_pmix["Item"] == "Hamachi Crudo"]
```
